# PlayStation Sales & Video Game Reviews Analysis
This notebook explores relationships between PlayStation sales data and game reviews.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ps_df = pd.read_csv("/mnt/data/PlayStation Sales and Metadata (PS3PS4PS5) (Oct 2025).csv")
vg_df = pd.read_csv("/mnt/data/video_game_reviews.csv")

ps_df.head(), vg_df.head()
```

```
ps_df.describe(include='all')
```

```
vg_df.describe(include='all')
```

```
merged = ps_df.merge(vg_df, left_on='Title', right_on='game', how='inner')
merged.head()
```

```
plt.figure(figsize=(12,6))
sns.heatmap(merged.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```
plt.figure(figsize=(10,5))
sns.scatterplot(data=merged, x='Sales', y='rating')
plt.title("Sales vs Review Rating")
plt.show()
```
