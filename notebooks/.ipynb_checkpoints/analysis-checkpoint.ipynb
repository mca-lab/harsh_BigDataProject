# Imported required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import data

```python
ps_df = pd.read_csv("../data/processed/PlayStation Sales and Metadata (PS3PS4PS5) (Oct 2025).csv")
vg_df = pd.read_csv("../data/processed/video_game_reviews.csv")
```

# Analysis on playstation Dataframe 

```python
sns.set_theme(style="whitegrid")

# 1) Distribution of Total Sales
plt.figure(figsize=(8,5))
sns.histplot(ps_df["Total Sales"], bins=30, kde=True)
plt.title("Distribution of Total Sales")
plt.xlabel("Total Sales (millions)")
plt.ylabel("Count")
plt.show()

# 2) Sales by Region
region_cols = ["NA Sales", "PAL Sales", "Japan Sales", "Other Sales"]
plt.figure(figsize=(8,5))
ps_df[region_cols].sum().plot(kind="bar")
plt.title("Total Sales by Region")
plt.ylabel("Total Units (millions)")
plt.show()

# 3) Metacritic vs Total Sales
plt.figure(figsize=(8,5))
sns.scatterplot(data=ps_df, x="metacritic", y="Total Sales", alpha=0.6)
plt.title("Metacritic Score vs Total Sales")
plt.xlabel("Metacritic")
plt.ylabel("Total Sales (millions)")
plt.show()

# 4) Correlation heatmap for numeric columns
plt.figure(figsize=(10,6))
corr = ps_df.select_dtypes("number").corr()
sns.heatmap(corr, annot=False, cmap="viridis")
plt.title("Correlation Heatmap – PlayStation Sales Dataset")
plt.show()
```

### Analysis on PlayStation Dataframe 

- General setup
    - Uses seaborn theme `whitegrid` for consistent styling and sets figure sizes for clarity.

- 1) Distribution of Total Sales
    - Histogram with KDE (`sns.histplot`) of `Total Sales` using 30 bins.
    - Purpose: inspect distribution, skewness, modes, and outliers.
    - Axis: Total Sales (millions) vs Count.

- 2) Sales by Region
    - Summed regional sales (`NA Sales`, `PAL Sales`, `Japan Sales`, `Other Sales`) plotted as a bar chart.
    - Purpose: compare total sales across regions.
    - Axis: Total Units (millions).

- 3) Metacritic vs Total Sales
    - Scatter plot (`sns.scatterplot`) of `metacritic` vs `Total Sales` with transparency (`alpha=0.6`).
    - Purpose: visualize relationship between critic score and sales.

- 4) Correlation heatmap for numeric columns
    - Compute correlation matrix for numeric columns and display via `sns.heatmap` (cmap="viridis").
    - Purpose: identify strong/weak correlations among numeric features.

# Analysis of VG sales data

```python
# 1) Distribution of User Rating
plt.figure(figsize=(8,5))
sns.histplot(vg_df["User Rating"], bins=20, kde=True)
plt.title("Distribution of User Ratings")
plt.xlabel("User Rating")
plt.ylabel("Count")
plt.show()

# 2) Price vs User Rating
plt.figure(figsize=(8,5))
sns.scatterplot(data=vg_df, x="Price", y="User Rating", alpha=0.4)
plt.title("Price vs User Rating")
plt.show()

# 3) Game Length vs User Rating
plt.figure(figsize=(8,5))
sns.scatterplot(data=vg_df, x="Game Length (Hours)", y="User Rating", alpha=0.4)
plt.title("Game Length vs User Rating")
plt.show()

# 4) Average rating by Genre (top 10 genres)
plt.figure(figsize=(10,5))
top_genres = (
    vg_df.groupby("Genre")["User Rating"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Genres by Average User Rating")
plt.ylabel("Avg User Rating")
plt.xlabel("Genre")
plt.show()
```

# Analysis of VG sales data

This cell produces four visualizations using the `vg_df` DataFrame to explore relationships between user ratings and other variables:

- **Distribution of User Ratings**
    - Histogram with KDE (`sns.histplot`) to show the spread and density of `User Rating`.
    - Purpose: inspect central tendency, skewness, and presence of multiple modes or outliers.

- **Price vs User Rating**
    - Scatter plot (`sns.scatterplot`) of `Price` vs `User Rating`.
    - Purpose: check for any relationship between a game's price and how users rate it.

- **Game Length vs User Rating**
    - Scatter plot of `Game Length (Hours)` vs `User Rating`.
    - Purpose: explore whether longer or shorter games tend to receive higher or lower user ratings.

- **Average Rating by Genre (Top 10)**
    - Bar plot of mean `User Rating` grouped by `Genre`, showing the top 10 genres by average rating.
    - Purpose: compare genres to identify which tend to receive higher user ratings; x-axis labels rotated for readability.

All plots set figure sizes for clarity and include titles and axis labels for interpretation.

# Analysis on merged dataset

```python
ps_df["Clean Game"] = ps_df["Game"].str.split("_", n=1).str[1]
ps_df["Clean Game"].head()
```

```python
merged = ps_df.merge(vg_df,left_on="Clean Game",right_on="Game Title",how="inner")
merged.info()
```

```python
merged.shape
```

```python
merged.describe(include="all")
```

```python
merged.dropna()
```

```python
merged.columns
```

```python
merged = merged.drop(["Game","Publisher_x","Total Shipped","Last Update","Requires Special Device","Clean Game","User Review Text"],axis=1)
```

```python
merged.columns
```

```python
merged.head()
```

```python
merged.info()
merged.describe(include='all')
```

```python
plt.figure(figsize=(12, 5))
sns.heatmap(merged.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
region_cols = ["NA Sales", "PAL Sales", "Japan Sales", "Other Sales"]

plt.figure(figsize=(8,4))
merged[region_cols].sum().plot(kind="pie")
plt.title("Total Sales by Region")
plt.show()
```

```python
merged.isnull().sum().sort_values(ascending=False)
```

```python
merged = merged.dropna()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=merged, x="User Rating", y="Total Sales", alpha=0.2)
plt.title("User Rating vs Total Sales")
plt.xlabel("User Rating")
plt.ylabel("Total Sales (Millions)")
plt.show()
```

 This cell creates a scatterplot showing the relationship between "User Rating" and "Total Sales".
 - Uses the cleaned 'merged' DataFrame.
 - plt.figure(figsize=(8,5)) sets the plot size.
 - sns.scatterplot(data=merged, x="User Rating", y="Total Sales", alpha=0.2) draws points with low opacity to reduce overplotting.
 - Title and axis labels clarify the axes (Total Sales are in millions).
 - plt.show() displays the plot.
 Purpose: visually inspect correlation, trends, and outliers between user ratings and sales.

```python
plt.figure(figsize=(12,5))
genre_sales = merged.groupby("Genre")["Total Sales"].mean().sort_values(ascending=False)
sns.barplot(x=genre_sales.index, y=genre_sales.values)
plt.xticks(rotation=90, ha="right")
plt.title("Average Sales by Genre")
plt.ylabel("Avg Sales (Millions)")
plt.show()
```

```python
plt.figure(figsize=(12,5))
genre_rating = merged.groupby("Genre")["User Rating"].mean().sort_values(ascending=False)
sns.barplot(x=genre_rating.index, y=genre_rating.values,width=0.5)
plt.xticks(rotation=45, ha="right")
plt.title("Average User Rating by Genre")
plt.ylabel("Avg User Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=merged, x="Price", y="User Rating")
plt.title("Price vs User Rating")
plt.show()
```
